#### Ejercicio 21. Realiza un trazado de las coordenadas (columnas latitud y longitud) del cinturón sur de Madrid por color de cada población (debes usar el subconjunto obtenido del Ejercicio 14) (★★★★)

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar un código simple que transforme las columnas de coordenadas en un diccionario de Python (agrega más información si es necesario) y agrégala al mapa.

```python
from ipyleaflet import Map, basemaps, Marker, AwesomeIcon

# Mapa centrado en (60 grados latitud y -2.2 grados longitud)
# Latitud, longitud
map = Map(center = (60, -2.2), zoom = 2, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenTopoMap)
map
```

```python
## Aquí: traza la coordenadas de los estados

## PON TU CÓDIGO AQUÍ:

diccionario_ej21= ds_cinturón_sur[['latitude','longitude','level5']].to_dict(orient = 'index')



small_icon = AwesomeIcon(icon='info-sign', icon_color='red', icon_size=(5, 5))
small_icon_2 = AwesomeIcon(icon='info-sign', icon_color='green', icon_size=(5, 5))
small_icon_3 = AwesomeIcon(icon='info-sign', icon_color='yellow', icon_size=(5, 5))
small_icon_4 = AwesomeIcon(icon='info-sign', icon_color='white', icon_size=(5, 5))

for Unnamed, valores in diccionario_ej21.items():
    latitude = float(valores['latitude'].replace(',', '.'))
    longitude = float(valores['longitude'].replace(',', '.'))
    level5 = valores['level5']  
    if level5 == 'Alcorcón':
        marcador = Marker(location=(latitude, longitude), icon = small_icon, draggable=True)
        map.add_layer(marcador)
    elif level5== 'Getafe':
        marcador = Marker(location=(latitude, longitude), icon = small_icon_2, draggable=True)
        map.add_layer(marcador)
    elif level5 == 'Fuenlabrada':
        marcador = Marker(location=(latitude, longitude), icon = small_icon_3, draggable=True)
        map.add_layer(marcador)
    elif level5 == 'Leganés':
        marcador = Marker(location=(latitude, longitude), icon = small_icon_4, draggable=True)
        map.add_layer(marcador)
        

    

map 
```
